# Experiment 1

## Removing an addictive substance does after a certain number of trials

```python
a1 = aa.AddictionAgent(max_time = 15, n_trials = 30, ETA=1, GAMMA=.5, rewards=[], substances=[(5,.5,8,(0,10))], verbose=False)
a1.run_TD()
```

```python
remove = aa.plot_comparison(a1.prediction_error, a1.value, "Removing Addictive Substance", 'Prediction Error', 'Value')
```

```python
remove.savefig('add_remove.png')
```

## Removing non addictive reward

```python
a2 = aa.AddictionAgent(max_time = 15, n_trials = 30, ETA=1, GAMMA=.5, rewards=[(5, 8, (0,10))], substances=[], verbose=False)
a2.run_TD()
```

```python
remove2 = aa.plot_comparison(a2.prediction_error, a2.value, "Removing Reward", 'Prediction Error', 'Value')
```

## Comparison
Removing an addictive substance of the same reward value as a non-addictive reward produces a larger dip in prediction error as well as a larger evaluation of value.

# Experiment 2

```python
cost_agent = aa.AddictionAgent(max_time = 15, n_trials = 30, ETA=1, GAMMA=.5, rewards=[(10, 4)], substances=[(5,.5,8, (0,5)),(2,.5,8, (5,30))], verbose=False)
cost_agent.run_TD()
```

```python
cost1 = aa.plot_comparison(cost_agent.prediction_error, cost_agent.value, "Decreasing Reward Component of Addiction", 'Prediction Error', 'Value')
```

Prediction error spike decreases when the reward component of the addictive substance decreases, but the value associated with the substance is still larger than just the reward component

```python
cost1.savefig('cost_exp.png')
```
