```python
import addiction_agent
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=3, linewidth=150)
```

```python
agent = addiction_agent.AddictionAgent(max_time = 10, n_trials = 10, ETA=1, GAMMA=1, rewards=[(10, 7)], substances=[], verbose=False)
```

```python
agent.run_TD()
```

```python
agent.value
```

```python
agent.plot_prediction_error(title = 'Prediction Error', savefig = 'baselinePE.png')
```

```python
agent.plot_value(title = "Values", savefig = 'baselineV.png')
```

```python
comp = addiction_agent.plot_comparison(agent.prediction_error, agent.value, "Baseline Prediction Errors and Values", 'Prediction Error', 'Values')
```

```python
comp.savefig('comp.png')
```
